```python
import pandas as pd
import joblib
import os

# /artifacts에서 모델과 테스트 데이터 로드
model = joblib.load('/artifacts/model.pkl')
test_df = pd.read_csv('/artifacts/mission15_test.csv')

print("test.csv 컬럼:", test_df.columns.tolist())  # 디버깅용

# 예측 (test.csv에는 목표변수 없으므로 전체 사용)
X_test = test_df
predictions = model.predict(X_test)

# result.csv로 저장 (첫 번째 컬럼을 ID로 가정)
result_df = test_df.iloc[:, [0]].copy()  # 첫 번째 컬럼을 ID로 사용
result_df['Performance Index'] = predictions.round()
result_df.to_csv('/artifacts/result.csv', index=False)

print("추론 완료! /artifacts/result.csv 확인하세요.")
print(f"샘플 예측값: {predictions[:5]}")
print(f"result.csv 저장됨. shape: {result_df.shape}")
```
